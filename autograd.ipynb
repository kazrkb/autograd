```python
def dy_dx(x):
    return 2*x
a = dy_dx(2)
print(a)
```

```python
import math

def dz_dx(x):
    return 2*x*math.cos(x**2)

print(dz_dx(2))
```

```python
import torch
```

```python
x = torch.tensor(3.0, requires_grad=True)
```

```python
y = x**2
```

```python
print(x)
print(y)
```

```python
y.backward()
```
